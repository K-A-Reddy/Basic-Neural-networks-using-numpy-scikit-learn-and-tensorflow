```python
from sklearn.datasets import fetch_openml
mnist=fetch_openml('mnist_784', version=1, as_frame=False)
```

```python
import numpy as np
X_train, X_test= mnist.data[60000:], mnist.data[:60000]
```

```python
#basic functions neccessary for our operations
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_delta(y,error):
   return  error * (y * (1 - y))  # sigmoid derivative

```

```python
# first thing we build is a neuron
class neuron:
    def __init__(self,in_size, activationfun,errorfun, learning_rate):
        self.errorfun=errorfun
        self.lrate=learning_rate
        self.insize=in_size
        self.weight=np.random.randn(in_size)*0.01
        self.bias= np.random.randn()*0.01
        self.activnfun=activationfun
    def forward(self,instance):
        self.instance=instance
        self.y=self.activnfun(np.dot(instance,self.weight)+self.bias)
        return self.y
    def backward(self,y_true):
        error = self.y - y_true
        delta=self.errorfun(self.y,error)
        # Update weights
        self.weight -= self.lrate * delta * self.instance
        # Update bias
        self.bias -= self.lrate * delta
        return delta

        
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]], float)
y = np.array([0,1,1,1], float)

neuron = neuron(2, sigmoid, sigmoid_delta, 0.1)

for epoch in range(2000):
    for i in range(len(X)):
        neuron.forward(X[i])
        neuron.backward(y[i])

print("Predictions:")
for i in range(len(X)):
    neuron.forward(X[i])
    print(X[i], "->", neuron.y)
```
